<br><h2 id="intro">ðŸ“ŒINTRODUCTION</h2>

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

<h2 id="intro">ðŸ“ŒIMPORTING NECESSARY LIBRARIES</h2>

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

<h2 id="intro">ðŸ“ŒLOAD THE PICTURE</h2>

```python
image = cv2.imread("/content/dog.jpg")
```

<h2 id="intro">ðŸ“ŒDISPLAY IMAGE USING OpenCV</h2>

```python
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h2 id="intro">ðŸ“ŒDISPLAY IMAGE USING MATPLOTLIB</h2>

```python
plt.imshow(image)
plt.axis(False)
plt.show()
```

<h2 id="intro">ðŸ“ŒCONVERT TO GREYSCALE IMAGE</h2>

```python
grey_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(grey_img)
```

<h2 id="intro">ðŸ“ŒINVERT THE IMAGE</h2>

```python
invert_image = cv2.bitwise_not(grey_img)
cv2_imshow(invert_image)
```

<h2 id="intro">ðŸ“ŒBLUR THE IMAGE</h2>

```python
blur_img = cv2.GaussianBlur(invert_image,(111,111),0)
cv2_imshow(blur_img)
```

<h2 id="intro">ðŸ“ŒINVERT THE BLUR IMAGE</h2>

```python
invblur_img=cv2.bitwise_not(blur_img)
cv2_imshow(invblur_img)
```

<h2 id="intro">ðŸ“ŒSKETCH<h2>

```python
sketch_img=cv2.divide(grey_img, invblur_img, scale=256.0)
```

```python
cv2.imwrite("save.png",sketch_img)
```

**DISPLAY SKETCH**

```python
cv2_imshow(sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

The Dog image is converted into pencil sketch with cv2 library in python.
